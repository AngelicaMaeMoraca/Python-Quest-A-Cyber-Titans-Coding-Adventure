## 3. String Operations

*Strings are text inside quotes `" "` or `' '`.  
We can **combine, slice, and modify** them.*

String operations are commonly used as methods in accessing string data type for data cleaning and debugging in some cases.

```python
# Define a sample string
sample_string = "Amazing world of capybaras"

# Length of a string
length = len(sample_string)
print(f"Length: {length}")
# Output: 27

# Accessing characters by index
first_char = sample_string[0]
last_char = sample_string[-1]
print(f"First character: {first_char}, Last character: {last_char}")
# Output: A, s

# Converting to uppercase
uppercase = sample_string.upper()
print(f"Uppercase: {uppercase}")
# Output: AMAZING WORLD OF CAPYBARAS

# Converting to lowercase
lowercase = sample_string.lower()
print(f"Lowercase: {lowercase}")
# Output: amazing world of capybaras

# Capitalizing the first character
capitalized = sample_string.capitalize()
print(f"Capitalized: {capitalized}")
# Output: Amazing world of capybaras

# Title case (capitalize the first character of each word)
title_case = sample_string.title()
print(f"Title Case: {title_case}")
# Output: Amazing World Of Capybaras

# Stripping whitespace
whitespace_string = "   Amazing world of capybaras   "
stripped = whitespace_string.strip()
print(f"Stripped: '{stripped}'")
# Output: 'Amazing world of capybaras'

# Splitting a string into a list
split_string = sample_string.split(" ")
print(f"Split: {split_string}")
# Output: ['Amazing', 'world', 'of', 'capybaras']

# Joining a list into a string
joined_string = " ".join(split_string)
print(f"Joined: {joined_string}")
# Output: Amazing world of capybaras

# Checking if a string starts with a substring
starts_with = sample_string.startswith("Amazing")
print(f"Starts with 'Amazing': {starts_with}")
# Output: True

# Checking if a string ends with a substring
ends_with = sample_string.endswith("capybaras")
print(f"Ends with 'capybaras': {ends_with}")
# Output: True

# Finding the index of a substring
index_of_of = sample_string.find("of")
print(f"Index of 'of': {index_of_of}")
# Output: 14

# Counting occurrences of a substring
count_of_a = sample_string.count("a")
print(f"Count of 'a': {count_of_a}")
# Output: 5

# String formatting
name = "Elon Musk"
age = 54
formatted_string = f"Name: {name}, Age: {age}"
print(f"Formatted String: {formatted_string}")
# Output: Name: Elon Musk, Age: 54

# Escaping special characters
escaped_string = "He said, \"Amazing world of capybaras\""
print(f"Escaped String: {escaped_string}")
# Output: He said, "Amazing world of capybaras"

# Reversing a string (using slicing)
reversed_string = sample_string[::-1]
print(f"Reversed String: {reversed_string}")
# Output: sarabypac fo dlrow gnizamA
```

### List, Tuple, and Set

1. **Lists**:
   - Ordered and mutable sequences.
   - Elements can be added, modified, or removed.
   - Defined using square brackets `[]`.

2. **Tuples**:
   - Ordered and immutable sequences.
   - Elements cannot be changed after creation.
   - Defined using parentheses `()`.

3. **Sets**:
   - Unordered collections of unique elements.
   - Elements cannot be duplicated.
   - Defined using curly braces `{}` or the `set()` function.
   - Supports operations like union, intersection, and difference.

### Common Operations

- **Lists**: Indexing, slicing, appending, removing, modifying elements.
- **Tuples**: Indexing, returning multiple values from functions, immutable.
- **Sets**: Adding, removing elements, handling duplicates, set operations (union, intersection, difference).

## 3.  Dictionaries
- Dictionaries offer efficient and flexible data storage, making them indispensable in many programming tasks.
- Understanding keys and values is essential for working effectively with dictionaries.
- Printing just keys or values can be useful for specific operations, such as iterating or accessing specific data.

## Importance of dictionaries
- Dictionaries allow efficient retrieval of values using unique keys.
- They are versatile data structures used for mappings, caching, and representing structured data.
- Dictionaries are commonly used in various domains, including web development, data processing, and system programming.

```python
# Creating a dictionary with student names and their corresponding grades
student_grades = {
    "Alice": 85,
    "Bob": 70,
    "Charlie": 95,
    "David": 60,
    "Emma": 75
}

# Printing the dictionary
print("Dictionary of student grades:", student_grades)

# Accessing values using keys
alice_grade = student_grades["Alice"]
print("\nAlice's grade:", alice_grade)

# Modifying values
student_grades["Bob"] = 80
print("\nUpdated dictionary after modifying Bob's grade:", student_grades)

# Adding new key-value pairs
student_grades["Frank"] = 90
print("\nUpdated dictionary after adding Frank's grade:", student_grades)

# Removing key-value pairs
removed_grade = student_grades.pop("David")
print("\nRemoved grade of David:", removed_grade)
print("Updated dictionary after removing David's grade:", student_grades)

# Printing just the keys
print("\nJust the keys:")
for name in student_grades.keys():
    print(name)

# Printing just the values
print("\nJust the values:")
for grade in student_grades.values():
    print(grade)

# Explanation of keys and values
print("\nKeys represent student names, and values represent their grades.")
print("Keys:", student_grades.keys())
print("Values:", student_grades.values())
```

## CSV Handling

1. Importing the pandas Library:

 - import pandas as pd
 - This imports the **pandas** library (commonly aliased as `pd`), which provides powerful tools for data manipulation and analysis.

2. Reading a CSV File:

 - data = pd.read_csv(csv_file_path)
 - This loads the CSV file into a **DataFrame** (a 2D labeled data structure similar to a spreadsheet).
 - The DataFrame is one of the most important data structures in data science, as it allows easy exploration, cleaning, and transformation of datasets.



3. Exploring the Dataset:

 - `print(data.head())` - Displays the first 5 rows of the dataset.
 -`print(data.info())` - Shows column names, non-null counts, and data types.
 -`print(data.describe())` - Provides summary statistics (mean, min, max, std) for numerical columns.


```python
# Import the pandas library
import pandas as pd

# Define the path to the CSV file
csv_file_path = "data.csv"

# Read the CSV file into a pandas DataFrame
data = pd.read_csv(csv_file_path)

# Print the entire CSV data (DataFrame)
print("Entire CSV data:")
print(data.to_string())  # to_string() ensures the full DataFrame is shown
```

Note: Be careful on the CSV file path. Wrong file path can lead to code error.

Now Let's use the 'grades' columns to demonstrate arithmetic operations

```python
# We can select these columns from the DataFrame
grades_df = data[['grades.DS101', 'grades.DS102', 'grades.DS103', 'grades.DS104', 'grades.DS105']]

print("Original grades data:")
display(grades_df)
```

```python
# Example 1: Calculate the sum of grades for each student across all courses
grades_sum = grades_df.sum(axis=1)
print("\nSum of grades for each student:")
display(grades_sum)
```

```python
# Example 2: Calculate the average grade for each student
grades_average = grades_df.mean(axis=1)
print("\nAverage grade for each student:")
display(grades_average)
```

```python
# Example 3: Find the minimum grade for each student
grades_min = grades_df.min(axis=1)
print("\nMinimum grade for each student:")
display(grades_min)
```

```python
# Example 4: Find the maximum grade for each student
grades_max = grades_df.max(axis=1)
print("\nMaximum grade for each student:")
display(grades_max)
```

```python
# Example 5: Add a constant to all grades (e.g., adding bonus points)
bonus_points = 5
grades_with_bonus = grades_df + bonus_points
print(f"\nGrades with {bonus_points} bonus points added:")
display(grades_with_bonus)
```

```python
# Example 6: Subtract a constant from all grades (e.g., adjusting for a curve)
curve_adjustment = 2
grades_after_curve = grades_df - curve_adjustment
print(f"\nGrades after subtracting {curve_adjustment} for a curve:")
display(grades_after_curve)
```

```python
# Example 7: Multiply grades by a factor (e.g., weighting)
weight_factor = 1.1
weighted_grades = grades_df * weight_factor
print(f"\nWeighted grades (multiplied by {weight_factor}):")
display(weighted_grades)
```

```python
# Example 8: Divide grades by a factor (e.g., normalizing)
normalization_factor = 100
normalized_grades = grades_df / normalization_factor
print(f"\nNormalized grades (divided by {normalization_factor}):")
display(normalized_grades)
```

## Graphing in Python
1. Importing the Required Library:

```python
import matplotlib.pyplot as plt
```

- This imports the Matplotlib library (commonly aliased as plt), which is the most widely used plotting library in Python.

- It allows us to create a wide variety of plots such as line plots, bar plots, and box plots.

2. Creating Example Data:

```python
Category = ['A', 'B', 'C', 'D', 'E']
X = [1, 4, 6, 8, 10]
Y = [5, 7, 2, 10, 7]
```

- We define simple lists of categories and numeric values. These will serve as the dataset for our plots.

3. Basic Plots

##### Line Plot

```python
plt.plot(X, Y, marker="o", linestyle="-", color="b")
plt.title("Line Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

- This will connect your (X, Y) points with a line.

- The circles (marker="o") mark each data point.

#### Bar Plot

```python
plt.bar(Category, Y, color="skyblue")
plt.title("Bar Plot Example")
plt.xlabel("category")
plt.ylabel("Y")
plt.show()
```

- Bars are drawn at each category in X-axis with height = Y.

#### Scatter Plot

Used to show the relationship between two variables.

```python
plt.scatter(X, Y, color="red", marker="o")
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

- Each point represents an (X, Y) pair. No line connects them.

#### Pie Chart

Used to show proportions or percentages.

```python
plt.pie(Y, labels=Category, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart Example")
plt.show()
```

#### Plot Customizations:

Add grid lines: plt.grid(True)

Rotate labels: plt.xticks(rotation=45)

Change figure size: plt.figure(figsize=(8,5))

Costumized lines syntax sample: plt.plot(X, Y, marker="o", linestyle="-", color="b", label="Y vs X")
